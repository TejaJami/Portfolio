```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("FuelConsumption.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df["Year"] = df["Year"].apply(str)
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(),  annot=True, cmap='coolwarm', fmt=".2f")
```

```python
sns.pairplot(df)
```

```python
sns.countplot(y ='MAKE', data = df,order=df['MAKE'].value_counts().index, palette = "Set2")
plt.title('Count of Vehicles by Make')
plt.xlabel('Count')
plt.ylabel('Make')
plt.show()
```

```python
sns.countplot(y ='VEHICLE CLASS', data = df,order=df['VEHICLE CLASS'].value_counts().index, palette = "Set2")
plt.title('Count of Vehicles by Vehicle Class')
plt.xlabel('Count')
plt.ylabel('Vehicle Class')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x="VEHICLE CLASS", y="ENGINE SIZE")
plt.title('Engine Size by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('Engine Size')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x="FUEL", y="ENGINE SIZE")
plt.title('Engine Size by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Engine Size')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x="TRANSMISSION", y="FUEL CONSUMPTION")
plt.title('Fuel Consumption by Transmission')
plt.xlabel('Transmission')
plt.ylabel('Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x="VEHICLE CLASS", y="FUEL CONSUMPTION")
plt.title('Fuel Consumption by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(19, 8))
sns.boxplot(data=df, x="MAKE", y="FUEL CONSUMPTION")
plt.title('Fuel Consumption by Make')
plt.xlabel('Make')
plt.ylabel('Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x="FUEL", y="FUEL CONSUMPTION")
plt.title('Fuel Consumption by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df, x="ENGINE SIZE", y="FUEL CONSUMPTION")
```

```python
Top10makers = df.groupby('MAKE')["FUEL CONSUMPTION"].mean().sort_values(ascending=False)[:10]
plt.figure(figsize=(10, 6))
sns.barplot(x=Top10makers.index, y=Top10makers.values, palette='coolwarm')
plt.title('Top 10 Mean Fuel Consumption by Make')
plt.xlabel('Make')
plt.ylabel('Mean Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
mean_fuel_consumption_by_engine_size = df.groupby('ENGINE SIZE')["FUEL CONSUMPTION"].mean().sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_fuel_consumption_by_engine_size.index, y=mean_fuel_consumption_by_engine_size.values, palette='coolwarm')
plt.title('Mean Fuel Consumption by Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Mean Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
mean_fuel_consumption_by_cylinders = df.groupby('CYLINDERS')["FUEL CONSUMPTION"].mean().sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_fuel_consumption_by_cylinders.index, y=mean_fuel_consumption_by_cylinders.values, palette='coolwarm')
plt.title('Mean Fuel Consumption by Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Mean Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
mean_fuel_consumption_by_transmission = df.groupby('TRANSMISSION')["FUEL CONSUMPTION"].mean().sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_fuel_consumption_by_transmission.index, y=mean_fuel_consumption_by_transmission.values, palette='coolwarm')
plt.title('Mean Fuel Consumption by Transmission')
plt.xlabel('Transmission')
plt.ylabel('Mean Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```
